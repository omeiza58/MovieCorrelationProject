```python
#importing libraries

import seaborn as sns
import numpy as np
import pandas as pd

#from pandas import *

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) #adjusting the configuration of the plots we will create

#import file

movies = pd.read_csv('movies.csv')
#viewing this file

movies.head()





```

```python
#showing percentage of null values present in a column

for col in movies.columns:
    pct_missing = np.mean(movies[col]. isnull())
    print('{} - {}%' .format(col, pct_missing))
```

```python
#DEAING WITH CELLS THAT CONTAIN MISSING VALUES

movies['rating'] = movies['rating'].replace(np.nan, 0)

movies['released'] = movies['released'].replace(np.nan, 0)

movies['score'] = movies['score'].replace(np.nan, 0)

movies['votes'] = movies['votes'].replace(np.nan, 0)

movies['writer'] = movies['writer'].replace(np.nan, 0)

movies['star'] = movies['star'].replace(np.nan, 0)

movies['country'] = movies['country'].replace(np.nan, 0)

movies['budget'] = movies['budget'].replace(np.nan, 0)

movies['gross'] = movies['gross'].replace(np.nan, 0)

movies['company'] = movies['company'].replace(np.nan, 0)

movies['runtime'] = movies['runtime'].replace(np.nan, 0)
```

```python
# from the above we see that five columns have float data types
#changing float data types to int64
#movies['rating'] = movies['rating'].astype('int64')


movies['votes'] = movies['votes'].astype('int64')

movies['budget'] = movies['budget'].astype('int64')

movies['gross'] = movies['gross'].astype('int64')

movies['runtime'] = movies['runtime'].astype('int64')

movies
```

```python
#LOOKING AT CORRELATION WITH COMPANY

movies_numerized = movies

for col_name in movies_numerized.columns:
    if(movies_numerized[col_name].dtype == 'object'): #if it has a column type with type object
        movies_numerized[col_name] = movies_numerized[col_name].astype('category')
        movies_numerized[col_name] = movies_numerized[col_name].cat.codes
movies_numerized
        
```
